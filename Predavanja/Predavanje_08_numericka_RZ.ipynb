### Danas radimo sledece: 

### 1) Pisemo numericko resenje jednacine prenosa zracenja. 

### 2) Plotujemo neka "zanimljiva" resenja i testiramo analiticka resenja Milneovog problema. 

### 3) Koristimo ga da iterativno resimo Milneov problem.

### 4) Vizualizujemo i diskutujemo jedan realisticni model zvezdane atmosfere koji je dobijen u RZ, ali ne-sivoj, ne-LTR atmosferi

```python
def formal_solution(tau, S, mu):
    
    ND = len(tau)
    I = np.zeros(ND)
    
    step = -1
    start = ND-1
    end = 0
    I[start] = S[start]
    
    if (mu < 0):
        step = 1
        start = 0 
        end = ND -1
        I[0] = 0.0
        
    for i in range(start+step, end+step, step):

        # Koristimo pretpostavku da nam je funkcija izvora konstantna na intervalu izmedju i i i-step
        S_mean = (S[i] + S[i-step]) * 0.5
        
        delta_tau = (tau[i-step] - tau[i]) / mu 
        
        I[i] = I[i-step] * np.exp(-delta_tau) + S_mean * (1.0 - np.exp(-delta_tau))
              
    return I
```

```python
# Log tau ima smisla zato sto pritisak opada logaritamski sa visinom.
# Pa neprozracnost opada logaritamski sa visinom
# Pa i opticka dubina (odnosno "razmaci" u optickoj dubini) opadaju logaritamski sa visinom
# Dakle log tau ~ -z

logtau = np.linspace(-3,3,61)
tau = 10.0 ** logtau
S = np.ones(61)

I_out = formal_solution(tau, S, 1)
I_in = formal_solution(tau, S, -1)
```

```python
plt.figure(figsize=[9,5.5])
plt.plot(logtau, S, label='S',linewidth = 4)
plt.plot(logtau, I_out, label = 'Izlazno I')
plt.plot(logtau, I_in, label = 'Ulazno I')
plt.xlabel("log $\\tau$")
plt.ylabel("Intenzitet")
plt.legend()
plt.tight_layout()
plt.savefig("FS_in_out.jpg",bbox_inches='tight')
```

### Ulazni intenzitet mozemo da shvatimo kao (ako je funkcija izvora konstantna):

### $I = S (1-e^{-\Delta \tau})$

### Za malo delta tau

### $I = S \Delta \tau$

### Ultimativno, mi zelimo da izracunamo J, za to nam treba integraljenje intenziteta po pravcima. 

```python
mu = np.array([-np.sqrt(0.6), 0 , np.sqrt(0.6)])
wmu = np.array([0.55555, 0.88888, 0.55555])

mu = mu / 2.0 
mu = mu+0.5

print ("diskretne vrednosti mu koje koristimo su: ", mu)
wmu = wmu / np.sum(wmu)
print ("tezine za te diskretne vrednosti su: ", wmu)
```

```python
J = np.zeros(len(S))
for m in range(0,3):
    I_in = formal_solution(tau,S, -mu[m])
    I_out = formal_solution(tau,S, mu[m])   
    J+= 0.5 * (I_in + I_out) * wmu[m]
```

```python
print (J)
```

```python
plt.figure(figsize=[9,5.5])
plt.plot(logtau, S, label='S',linewidth = 4)
plt.plot(logtau, J, label='J = Lambda [S]', linewidth = 3)
plt.ylim([0.0,1.05])
plt.xlabel("log $\\tau$")
plt.ylabel("J")
plt.legend()
plt.tight_layout()
plt.savefig("J_isothermal.jpg",bbox_inches='tight')
```
