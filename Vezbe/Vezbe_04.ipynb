```python
import numpy as np
import matplotlib.pyplot 
```

```python
# zelimo da isplotujemo eksponencijalne integrale koji se pojavljuju u 
# Schwartzshildovim operatorima. Oni su neke specijalne funkcije
```

```python
from scipy.special import expn
```

```python
# Definisimo neko S na nekom opsegu tau

t = np.linspace(0,5,101)

# napravimo neku linearnu funkciju (ne mora da bude linearna, ovo je samo za
# demonstraciju)
F = 1
S = 3./4. * F * t + F / 2.
```

```python
# Recimo da zelimo da preko eksponencijalnog integrala, tj. lambda operatora
# vidimo kako bi izgledalo J od nekog tau
```

```python
tau = 1.0
```

### Znamo da:

## $J(\tau) = \int_0^{\infty} S(t) E_1(|\tau-t|) dt$

```python
plt.figure(figsize=[10,7])
plt.plot(t,S, label='Funkcija izvora')
plt.plot(t,expn(1, np.abs(tau-t)), label = 'Kernel (E1)')
plt.plot(t,expn(1, np.abs(tau-t)) * S, label = 'Podintegralna funkcija')
plt.legend()
plt.xlabel("$\\tau$")
plt.ylabel("S")
```

## Za fluks imamo: 

## $H(\tau) = \int_\tau^{\infty} S(t) E_2(t-\tau) dt - \int_0^{\tau} S(t) E_2(\tau-t) dt$

```python
plt.figure(figsize=[10,7])
plt.plot(t,S, label='Funkcija izvora')
plt.plot(t,expn(2, np.abs(tau-t)), label = 'Kernel (E2)')
plt.plot(t,expn(2, np.abs(tau-t)) * S, label = 'Podintegralna funkcija')
plt.legend()
plt.xlabel("$\\tau$")
plt.ylabel("S")
```

## Ovo F mora da uvek pozitivno (tacnije, nije F nego H, ali ok). Sto prakticno znaci da S mora da opada ka povrsini, jer je kernel i dalje manje vise simetrican.

```python
# Hajde da isplotujemo exp integrale za t (sto je recimo x)
```

```python
plt.figure(figsize=[10,7])
plt.plot(t,np.exp(-t), label='exp(-t)')
plt.plot(t,expn(1,t), label='E1')
plt.plot(t,expn(2,t), label='E2')
plt.plot(t,expn(3,t), label='E3')
plt.legend()
plt.xlabel("$\\tau$")
plt.ylabel("S")
```
