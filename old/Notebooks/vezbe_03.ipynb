### Treci zadatak

```python
a = 0.5 
b = 0.5
```

```python
# Kako da definisemo r_lambda?

# Modelujemo oblike spektralne linije

# Lokalno, r_lambda ce biti nesto sto zovemo apsorpcioni profil! 
```

```python
# Modelujemo apsorpcioni profil kao 1+ neki gausijan (1 jer uvek imamo kontinuum)
```

```python
# Gausijan:
A = 100.0
llambda_0 = 656.28 # nm, H alfa
delta_llambda = 0.3 # nm, lupamo (za nekoliko casova cemo znati bolje)

llambda = np.linspace(-5*delta_llambda, 5*delta_llambda,201) + llambda_0

# Gausijan, normiran da mu maksimum bude A

gausijan = A * np.exp(-((llambda-llambda_0)/delta_llambda)**2.0)

r_lambda = 1.0 + gausijan
```

```python
# Ovo ce sada biti plot koji nam govori kako nasa atmosfera apsorbuje na nekom
# opsegu talasnih duzina

plt.figure(figsize=[10,6])
plt.plot(llambda, r_lambda)
plt.xlabel("Talasna duzina")
plt.ylabel("$r_\lambda$")
```

```python
# Ocekujemo da ce izlazni intenzitet biti ovo ISTO samo obrnuto. Hajde da vidimo:
```

```python
I_lambda = a + b / r_lambda
```

```python
plt.figure(figsize=[13,6])
plt.subplot(121)
plt.plot(llambda, r_lambda)
plt.xlabel("Talasna duzina")
plt.ylabel("$r_\lambda$")
plt.subplot(122)
plt.plot(llambda, I_lambda)
plt.xlabel("Talasna duzina")
plt.ylabel("$I_\lambda$")
```
