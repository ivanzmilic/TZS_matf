Kako izlazni specificni intenzitet zavisi od ukupne opticke dubine objekta i ulaznog intenziteta?

```python
I_0 = 1.0 

# Pp da je funkcija izvora manja od ulaznog intenziteta 

S = 10.0 

# Uzmimo razne vrednosti tau_max 

logtau_max = np.linspace(-3,5,71)
tau_max = 10.0 ** logtau_max

#print (tau_max) #sanity check
```

```python
I_plus = I_0 * np.exp(-tau_max) + S * (1.0 - np.exp(-tau_max))
```

```python
plt.figure(figsize=[8,5])
plt.semilogx(tau_max, I_plus)
plt.xlabel("Ukupna opticka dubina")
plt.ylabel("Izlazni intenzitet")
```

Sada pokusavamo da pretpostavimo neku zavisnost koeficijenta apsorpcije od talasne duzine, pa da vidimo kako ta zavisnost utice na spektar objekta. Pretpostavicemo fiksan granicni uslov za sve talasne duzine (I_0 = 1), i konstantnu funciju izvora.

```python
I_0 = 1.0
S = 0.3

llambda = np.linspace(200, 1000, 801) # u nm

chi_llambda = np.ones(801)

# Zamislimo da imamo neki skok u neprozracnosti npr na 397 nm (otprilike
# jonizacija vodonika sa n=2).

chi_llambda[:197] *= 10.0

delta_l = 1.0

tau_llambda = chi_llambda * delta_l
```

```python
# Isplotujmo koeficijent apsorpcije:

plt.figure(figsize=[8,5])
plt.plot(llambda, chi_llambda)
plt.xlabel("Talasna duzna")
plt.ylabel("Koeficijent apsorpcije")
```

```python
I_plus = I_0 * np.exp(-tau_llambda) + S*(1.0 - np.exp(-tau_llambda))
```

```python
# Ovo je, u stvari, nas izlazni spektar!

plt.figure(figsize=[8,5])
plt.plot(llambda, I_plus)
plt.xlabel("Talasna duzna")
plt.ylabel("Izlazni intenzitet")
```
